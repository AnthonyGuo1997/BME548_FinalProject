```python
class alex_model(tf.keras.Model):
    def __init__(self):
        super().__init__()
        # define the layers
        
        # block 1
        self.conv1 = tf.keras.layers.Conv2D(
        filters = 96,
        kernel_size = [11, 11],
        strides=(4,4),
        activation =tf.nn.leaky_relu
        )
        
        self.norm1 = tf.keras.layers.BatchNormalization()

        self.pool1 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=(2,2))
        
        #block 2
        self.conv2 = tf.keras.layers.Conv2D(
        filters = 256,
        kernel_size = [5, 5],
        strides=(1,1),
        activation =tf.nn.leaky_relu
        )
        
        self.norm2 = tf.keras.layers.BatchNormalization()

        self.pool2 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=(2,2))
        
        #block 3
        self.conv3 = tf.keras.layers.Conv2D(
        filters = 384,
        kernel_size = [3, 3],
        strides=(1,1),
        activation =tf.nn.leaky_relu
        )
        
        self.norm3 = tf.keras.layers.BatchNormalization()

        
        #block 4
        self.conv4 = tf.keras.layers.Conv2D(
        filters = 384,
        kernel_size = [3, 3],
        strides=(1,1),
        activation =tf.nn.leaky_relu
        )
        
        self.norm4 = tf.keras.layers.BatchNormalization()
        
        #block 5
        self.conv5 = tf.keras.layers.Conv2D(
        filters = 256,
        kernel_size = [3, 3],
        strides=(1,1),
        activation =tf.nn.leaky_relu
        )
        self.norm5 = tf.keras.layers.BatchNormalization()
        self.pool5 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=(2,2))
    
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(units=4096, activation = tf.nn.leaky_relu)
        self.drop1 = tf.keras.layers.Dropout(0.5)
        self.dense2 = tf.keras.layers.Dense(units=4096, activation = tf.nn.leaky_relu)
        self.drop2 = tf.keras.layers.Dropout(0.5)
        self.dense3 = tf.keras.layers.Dense(units=2)
        
    def call(self, inputs):
        # define the network structure
        x = self.conv1(inputs)
        x = self.norm1(x)
        x = self.pool1(x)
        
        x = self.conv2(x)
        x = self.norm2(x)
        x = self.pool2(x)
        
        x = self.conv3(x)
        x = self.norm3(x)
                
        x = self.conv4(x)
        x = self.norm4(x) 
                
        x = self.conv5(x)
        x = self.norm5(x)
        x = self.pool5(x)
        
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.drop1(x)
        x = self.dense2(x)
        x = self.drop2(x)
        x = self.dense3(x)
        return x
model = alex_model()
```

```python
class VGG16_model(tf.keras.Model):
    def __init__(self):
        super().__init__()
        # define the layers
        
        # block 1
        self.conv1_1 = tf.keras.layers.Conv2D(
        filters = 64,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv1_2 = tf.keras.layers.Conv2D(
        filters = 64,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.pool1 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        
        #block 2
        self.conv2_1 = tf.keras.layers.Conv2D(
        filters = 128,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv2_2 = tf.keras.layers.Conv2D(
        filters = 128,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.pool2 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        
        #block 3
        self.conv3_1 = tf.keras.layers.Conv2D(
        filters = 256,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv3_2 = tf.keras.layers.Conv2D(
        filters = 256,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv3_3 = tf.keras.layers.Conv2D(
        filters = 256,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.pool3 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        
        #block 4
        self.conv4_1 = tf.keras.layers.Conv2D(
        filters = 512,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv4_2 = tf.keras.layers.Conv2D(
        filters = 512,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv4_3 = tf.keras.layers.Conv2D(
        filters = 512,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        # step  pooling
        self.pool4 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        
        #block 5
        self.conv5_1 = tf.keras.layers.Conv2D(
        filters = 512,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv5_2 = tf.keras.layers.Conv2D(
        filters = 512,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        self.conv5_3 = tf.keras.layers.Conv2D(
        filters = 512,
        kernel_size = [3, 3],
        padding ='same',
        activation =tf.nn.relu
        )
        # step  pooling
        self.pool5 = tf.keras.layers.MaxPool2D(pool_size=[2, 2], strides=2)
        
    
        self.flatten = tf.keras.layers.Flatten()
        self.dense1 = tf.keras.layers.Dense(units=4096, activation = 'relu')
        self.dense2 = tf.keras.layers.Dense(units=4096, activation = 'relu')
        self.dense3 = tf.keras.layers.Dense(units=4)
        
    def call(self, inputs):
        # define the network structure
        x = self.conv1_1(inputs)
        x = self.conv1_2(x)
        x = self.pool1(x)
        
        x = self.conv2_1(x)
        x = self.conv2_2(x)
        x = self.pool2(x)
        
        x = self.conv3_1(x)
        x = self.conv3_2(x)
        x = self.conv3_3(x)
        x = self.pool3(x)
                
        x = self.conv4_1(x)
        x = self.conv4_2(x)
        x = self.conv4_3(x)
        x = self.pool4(x) 
                
        x = self.conv5_1(x)
        x = self.conv5_2(x)
        x = self.conv5_3(x)
        x = self.pool5(x)
        
        x = self.flatten(x)
        x = self.dense1(x)
        x = self.dense2(x)
        x = self.dense3(x)
        return x
model = VGG16_model()
```
